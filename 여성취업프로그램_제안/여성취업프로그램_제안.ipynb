#### T검정
T Score

- t-score가 크면 두 그룹이 다르다는 것을 의미한다.

- t-score가 작으면 두 그룹이 비슷하다는 것을 의미한다.

​

P-values

- p-value는 5%수준에서 0.05이다.

- p-values는 작은 것이 좋다. 이것은 데이터가 우연히 발생한 것이 아니라는 것을 의미한다.

- 예를 들어 p-value가 0.01 이라는 것은 결과가 우연히 나올 확률이 1%에 불과하다는 것을 의미한다.

- 대부분의 경우 0.05 (5%) 수준의 p-value를 기준으로 삼는다. 이 경우 통계적으로 유의하다고 한다.

\

결과값에서 e의 의미
 - e-16 = 10의 -16제곱 (즉, 자릿수) 
 - 예를 들어 1e-3 = 0.001
 - 6.746e-16 = 0.000000000000006746

```python
df_t = df[df['시도별'] == '계']
```

```python
df_t
```

```python
df_t = df_t[df_t['성별'] != '계']
```

```python
del df_t['시도별']
```

```python
df_t.columns = ['시점', '성별', '15세이상인구', '경제활동인구', '비경제활동인구',
                '경제활동참가율', '실업률', '고용률', '15~64세 고용률']
```

```python
# 성별로 데이터 나누기
m = df_t.loc[df_t['성별'] == '남자']
f = df_t.loc[df_t['성별'] == '여자']
```

- 경제활동참가율 T검정

```python
from scipy.stats import bartlett
bartlett(m['경제활동참가율'], f['경제활동참가율'])

# p값이 0.05보다 크기 때문에 등분산성이 있다고 판단 T검정 실시
# 등분산성이 없다면 ttest_ind()에는 equal_var=False 옵션을 추가
```

```python
from scipy.stats import ttest_ind
ttest_ind(m['경제활동참가율'], f['경제활동참가율'])

# 귀무가설 : 성별 경제활동참가율의 평균이 차이가 없다
# 대립가설 : 성별 경제활동참가율의 평균이 차이가 있다.

# pvalue = 1.0477592183837652e-16
# p값이 0.05보다 작기 때문에 귀무가설 기각, 대립가설 채택
# 즉, 남자 고용률이 여자의 고용률보다 차이가 있다.

# e-16 = 10의 -16제곱 (즉, 자릿수) 
# 예를 들어 1e-3 = 0.001
# 6.746e-16 = 0.000000000000006746
```

```python
m['경제활동참가율'].mean() - f['경제활동참가율'].mean()  # 20.042857142857144 정도의 차이를 보인다.
```

- 고용률 T검정

```python
print(bartlett(m['고용률'], f['고용률']))   
# pvalue=0.8705811383162056 
# p값이 0.05보다 크기 때문에 등분산성이 있다고 판단

print(ttest_ind(m['고용률'], f['고용률']))  
# pvalue=1.3562317002581606e-16
# T검정 결과, p값이 0.05보다 작기 때문에 귀무가설 기각, 대립가설 채택
# 귀무가설 : 성별 고용률 평균의 차이가 없다.
# 대립가설 : 성별 고용률 평균의 차이가 있다.
# p값이 0.05보다 작기 때문에 두 집단 간의 평균이 통계적으로 유의미한 차이가 존재한다.

m['고용률'].mean() - f['고용률'].mean()   # 19.414285714285725 정도의 차이가 존재.
```

- 15~64세 고용률 T검정

```python
print(bartlett(m['15~64세 고용률'], f['15~64세 고용률']))   
# pvalue=0.3971172130623736
# p값이 0.05보다 크기 때문에 등분산성이 있다고 판단

print(ttest_ind(m['15~64세 고용률'], f['15~64세 고용률']))  
# pvalue=7.350477590670831e-16
# T검정 결과, p값이 0.05보다 작기 때문에 귀무가설 기각, 대립가설 채택
# 귀무가설 : 성별 15~64세 고용률 평균의 차이가 없다.
# 대립가설 : 성별 15~64세 고용률 평균의 차이가 있다.
# p값이 0.05보다 작기 때문에 두 집단 간의 평균이 통계적으로 유의미한 차이가 존재한다.

m['15~64세 고용률'].mean() - f['15~64세 고용률'].mean()   # 18.042857142857137 정도의 차이가 존재.
```

- 실업률 T검정?

```python
print(bartlett(m['실업률'], f['실업률']))   
# pvalue=0.17677337548837152
# p값이 0.05보다 크기 때문에 등분산성이 있다고 판단

print(ttest_ind(m['실업률'], f['실업률']))  
# pvalue=0.39021599306449295
# T검정 결과, p값이 0.05보다 크기 때문에 귀무가설 채택, 대립가설 기각
# 귀무가설 : 성별 실업률 평균의 차이가 없다.
# 대립가설 : 성별 실업률 평균의 차이가 있다.
# p값이 0.05보다 크기 때문에 두 집단 간의 평균이 통계적으로 유의미한 차이가 존재하지 않음.

m['실업률'].mean() - f['실업률'].mean()   # -0.1857142857142855
```

###### 불용어 제거

```python
count_len = 10
rank_txt = {}

#items를 써야함/키,벨류 값 나와야하니깐
for key,value in count.items():
    if value > count_len:
        rank_txt[key] = value
        
rank_txt
```

```python
# 위에 count.most_common() 실행해서 불용어 찾아서 제거
stopword = ['개강','과정','유료','급','기','및','반','비대','신규','위','자격',
            '전문', '터','회','오전','전반','차','초','월','통합','야간','카드']

# 불용어 제거
[rank_txt.pop(key) for key in stopword]
# 실행 후 숫자들이 나오면 제대로 실행됨. (숫자는 인덱스 느낌??)
```

#### 서울시 산업체현황 종사자수 (산업대분류별, 동별, 성별) 통계

```python
filename = '/content/서울시 사업체현황 종사자수(산업대분류별동별성별) 통계.xlsx'
```

```python
#csv파일 읽어오기
#df = pd.read_csv(filename, encoding='cp949')
df7 = pd.read_excel(filename, na_values = ['-'])  #엑셀파일
                             # NA 처리 문자 지정
```

```python
df7.shape  # 크기 확인
```

```python
df7.isna().sum()  # 결측값 확인
```

- 컬럼명 해석

1. 합계
 - 합계 - 사업체수
 - 합계.1 - 총종사자수
 - 합계.2 - 여성종사자수

 \
2. 다른 컬럼들
 - 그냥 이름 - 사업체 수
 - 이름.1 - 여성종사자수


```python
df7.head()
```

```python
df7 = df7.fillna(0)
```

```python
df7.iloc[:, 6:] = df7.iloc[:, 6:].astype(int)
```

```python
gang = df7[df7['자치구'] == '강서구']
```

```python
gang.head()
```

```python
gang.info()
```

```python
# 취업자수, 증가율 컬럼 나누기
idx = [x for x in range(6, 44, 2)]
id = [x for x in range(7, 44, 2)]
```

```python
# 기본 컬럼들 추가하기
for x in [0, 2]:
    idx.append(x)
    id.append(x)
print(idx)
print(id)
```

```python
total = gang.iloc[:, idx]
worker = gang.iloc[:, id]
```

```python
total = total[total['동'] == '소계']
worker = worker[worker['동'] == '소계']
```

```python
total.iloc[:, :-1]
```

```python
worker.columns = total.columns
```

```python
# 데이터 행렬바꾸기 (R에서의 reshape2 패키지와 동일)
total = pd.melt(total.iloc[:, :-1], id_vars = ['기간'])
worker = pd.melt(worker.iloc[:, :-1], id_vars = ['기간'])
```

```python
total.columns = ['기간', '구분', '값']
worker.columns = ['기간', '구분', '값']
```

```python
# 기간의 값을 '...년'으로 바꾸기 위해 카테고리형으로 바꾸기
total['기간'] = total['기간'].astype('category')
worker['기간'] = worker['기간'].astype('category')
```

```python
# '기간의 값을 '...년'으로 바꾸기
total['기간'].cat.categories = ['2017년', '2018년', '2019년']
worker['기간'].cat.categories = ['2017년', '2018년', '2019년']
```

```python
total.sort_values(by = ['기간', '값'], ascending = [False, False], inplace = True)
worker.sort_values(by = ['기간', '값'], ascending = [False, False], inplace = True)
```

```python
fig = plt.figure(figsize = (20, 9))   # 그래프 화면 크기

g = sns.barplot(x = '직업명', y = '값', hue= '시기', data = employ, ci = None, palette = "colorblind", alpha = 0.5)  # "Spectral" , "coolwarm"

# 제목
plt.title('취업자 증가 속도가 빠른 직업', fontsize= 25)

# 축제목
plt.xlabel('')
plt.ylabel('')

# x눈금 이름
plt.xticks(fontsize = 14, rotation = 70)

# 범례 위치
plt.legend(loc='upper right', fontsize= 15)

plt.show()
```

```python
indust.columns = ['산업명', '2015년', '2020년', '2025년', '2030년']
grow.columns = ['산업명', '2015년 ~ 2020년', '2020년 ~ 2025년', '2025년 ~ 2030년', '2020년 ~ 2030년']
```

```python
# 데이터 행렬바꾸기 (R에서의 reshape2 패키지와 동일)
indust = pd.melt(indust, id_vars = ['산업명'])
grow = pd.melt(grow, id_vars = ['산업명'])
```

```python
grow['값'] = grow['값'].astype('float64')
```

```python
fig = plt.figure(figsize = (23, 8))   # 그래프 화면 크기

g = sns.barplot(x = '산업명', y = '값', hue= '시기', data = indust, ci = None, palette = "Set1", alpha = 0.5)  # "Spectral" , "coolwarm"

# 제목
plt.title('취업자 증가 속도가 빠른 산업', fontsize= 25)

# 축제목
plt.xlabel('')
plt.ylabel('')

# x눈금 이름
plt.xticks(fontsize = 14, rotation = 80)

# 범례 위치
plt.legend(loc='upper right', fontsize= 18) # , bbox_to_anchor=(1, 0.5)

plt.show()
```

- 연평균 증가율

```python
fig = plt.figure(figsize = (9, 14))   # 그래프 화면 크기

sns.scatterplot(y = "산업명", x = "값", hue = '시기',
                data = grow, size = "값",
                palette = 'spring', 
                legend = True,   # 버블의 크기 범례를 나타남
                sizes = (20, 400),  # 버블의 크기
                alpha = 0.5)  # 버블의 투명도


# 제목
plt.title('취업자 증가 속도가 빠른 산업 (연평균 증가율)', fontsize= 20)

# 축제목
plt.xlabel('')
plt.ylabel('')

# x눈금 이름
plt.yticks(fontsize = 14)

# 범례 위치
plt.legend(loc='center right', bbox_to_anchor=(1.35, 0.5), fontsize= 14)

plt.show()
```
